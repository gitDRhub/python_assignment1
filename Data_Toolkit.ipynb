**2.Explain the following termsg**
* The difference in np.array, np.asarray and np.asanyarrayX
* The difference between Deep copy and shallow copyX

**Ans.**Difference between np.array, np.asarray, and np.asanyarray:
These are all NumPy functions that convert input data (like lists or tuples) into arrays, but they behave slightly differently in certain cases.

1. np.array()

1.Always creates a new array (copy by default).

2.You can force it not to copy with copy=False, but it still makes a copy unless the input is already an array with the exact same dtype.

```python
a = [1, 2]
arr = np.array(a)
```

2. np.asarray()

1.Avoids copying if the input is already a NumPy array (with the right dtype).

2.It's more efficient when you're not sure if the input is already an array.

```python
a = np.array([1, 2])
b = np.asarray(a)  # No copy made
```

3. np.asanyarray()

1.Like asarray(), but it preserves subclasses (e.g., matrix, masked arrays).

2.Useful when working with custom array types.




```python
from numpy import matrix
m = matrix([[1, 2], [3, 4]])
a = np.asanyarray(m)
```

Difference Between Deep Copy and Shallow Copy:
This is about how data is copied, especially with mutable structures like arrays and lists.

1. Shallow Copy:

Creates a new object, but references the same data.

Changes to nested or referenced objects affect both copies.

```python
import copy

a = [1, [2, 3]]
b = copy.copy(a)
b[1][0] = 99
print(a)  # [1, [99, 3]] — inner list changed in both
```

2. Deep Copy:

Recursively copies everything, including nested structures.

Changes in the copied object don’t affect the original.


```python
c = copy.deepcopy(a)
c[1][0] = 100
print(a)  # [1, [99, 3]] — unchanged
```

With NumPy:

```python
import numpy as np

x = np.array([1, 2, 3])
y = x.copy()  # Deep copy
z = x         # Shallow (just reference)

y[0] = 99
z[1] = 88

print(x)  # x is changed by z but not by y
```

**14.Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create time-series data
dates = pd.date_range(start='2023-01-01', periods=30, freq='D')
temperature = np.random.uniform(15, 35, size=30)   # Random temperatures between 15°C and 35°C
humidity = np.random.uniform(30, 90, size=30)      # Random humidity between 30% and 90%

# Create DataFrame
weather_data = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# Temperature line (left y-axis)
ax1.plot(weather_data['Date'], weather_data['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Humidity line (right y-axis)
ax2 = ax1.twinx()
ax2.plot(weather_data['Date'], weather_data['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.tight_layout()
plt.show()
```

**15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:

Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create the histogram
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogram')

# Compute the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, loc=np.mean(data), scale=np.std(data))

# Overlay the PDF line
plt.plot(x, pdf, 'r', linewidth=2, label='PDF')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Show legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

**17.g Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
x = np.random.randn(200)
y = np.random.randn(200)

# Determine quadrant for each point
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    elif x > 0 and y < 0:
        return 'Quadrant IV'
    else:
        return 'Origin'

# Create DataFrame
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set2')

# Add formatting
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.tight_layout()

# Show plot
plt.show()
```

**18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Show plots in notebook (optional: use output_file('sine_wave.html') to save as HTML)
output_notebook()

# Generate x and y values for sine wave
x = np.linspace(0, 4 * np.pi, 500)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis', width=700, height=400)

# Add line plot
p.line(x, y, line_width=2, color='navy', legend_label='sin(x)')

# Grid lines are enabled by default in Bokeh
# You can customize them like this if desired:
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

**19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral11

# Show plots in notebook (optional: use output_file('bar_chart.html') to save as HTML)
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
values = np.random.randint(10, 100, size=len(categories))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a Bokeh figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, width=800, height=400)

# Add bars
p.vbar(x='categories', top='values', source=source, width=0.9, color=Spectral11[:len(categories)], legend_field='categories')

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Category"
p.yaxis.axis_label = "Value"

# Remove gridlines and axis lines for aesthetics
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Show legend and display plot
p.legend.title = 'Category'
p.legend.location = 'top_left'
p.legend.orientation = 'horizontal'

# Show the plot
show(p)
```

**19.Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.rand(100) * 10  # Random values scaled by 10

# Create a Plotly figure
fig = go.Figure()

# Add line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Line'))

# Set title and axis labels
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show plot
fig.show()
```
