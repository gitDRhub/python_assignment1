5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
size = 100

# Create correlated data
x = np.random.normal(50, 10, size)
y = x * 0.5 + np.random.normal(5, 2, size)  # Positively correlated with x
z = -0.7 * x + np.random.normal(10, 3, size)  # Negatively correlated with x

# Construct DataFrame
data = pd.DataFrame({'Feature X': x, 'Feature Y': y, 'Feature Z': z})

# Calculate correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.


```python
import pandas as pd
import numpy as np
import plotly.express as px

# Generate sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot table for heatmap
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create heatmap using Plotly
fig = px.imshow(sales_pivot, text_auto=True, aspect="auto", title="Variation in Sales Across Different Months and Days",
                labels=dict(x="Month", y="Day", color="Sales"))
fig.show()
```

 Using the given x and y data, generate a 3D surface plot to visualize the function

```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays to create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot using Plotly
fig = go.Figure(data=[
    go.Surface(z=z, x=x, y=y, colorscale='Viridis')
])

fig.update_layout(
    title="3D Surface Plot of f(x, y) = sin(sqrt(x^2 + y^2))",
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
from bokeh.plotting import figure, output_notebook, show
import numpy as np

# Prepare the output in the notebook
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.




```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data for x and y, as well as sizes and colors for the markers
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.randint(10, 30, 100)
colors = np.random.choice(['blue', 'green', 'red', 'purple'], 100)

# Create a data source
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a scatter plot
p = figure(title="Random Scatter Plot", width=700, height=400)
p.scatter('x', 'y', size='sizes', color='colors', source=source, fill_alpha=0.6)

# Show the plot
show(p)
```

 3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
# Updated data for the bar chart
from bokeh.models import FactorRange # Import FactorRange from bokeh.models
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a bar chart
p = figure(x_range=FactorRange(*fruits), title="Fruit Counts", width=700, height=400)
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="coral")

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.


```python
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.layouts import layout
import numpy as np

# Data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for Bokeh's ColumnDataSource
heatmap_data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'values': data_heatmap.flatten()
}

# Color mapper for the heatmap
color_mapper = LinearColorMapper(palette="Viridis256", low=heatmap_data['values'].min(), high=heatmap_data['values'].max())

# Create the heatmap plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=600)
p.rect(x='x', y='y', width=1/10, height=1/10, source=heatmap_data, fill_color=transform('values', color_mapper), line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0), title="Values")
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
